```python
import pandas as pd
import random
from datetime import datetime, timedelta
import string

# Function to generate random release versions in the format X.Y.Z
def generate_random_version():
    return f"{random.randint(1, 9)}.{random.randint(0, 9)}.{random.randint(0, 9)}"

# Define the number of rows
num_rows = 10000

# Generate lists of possible NE_Names and Board_Types
names = [f"NE_{letter}" for letter in string.ascii_uppercase]  # NE_A to NE_Z
board_types = [f"Type_{letter}" for letter in string.ascii_uppercase]  # Type_A to Type_Z
types = ["Router", "Switch", "Firewall", "Gateway", "Hub"]
versions = ["1.0", "2.1", "3.0", "4.5", "5.0"]


#---------------------------------

#Network

# Function to generate random IP addresses
def generate_ip_address(base):
    return f"{base}.{random.randint(1, 254)}"

# Function to generate random configuration dates
def generate_random_date(start_date, num_days):
    random_days = random.randint(0, num_days)
    return (start_date + timedelta(days=random_days)).strftime("%Y-%m-%d")

# Define the number of rows
num_rows = 10000

# Base IP for generation
base_ip = "192.168"


#--------------------------------

import pandas as pd
import random
from datetime import datetime, timedelta

# Function to generate random locations
def generate_location():
    locations = [
        "New York", "San Francisco", "London", "Berlin", "Tokyo", "Mumbai", "Chennai", "Bangalore", "Manhattan", "Beijing",
        "Los Angeles", "Chicago", "Houston", "Phoenix", "Philadelphia", "Dallas", "San Diego", "San Jose", "Austin", "Jacksonville",
        "Fort Worth", "Columbus", "Charlotte", "San Francisco", "Indianapolis", "Seattle", "Denver", "Washington", "Boston", "El Paso",
        "Nashville", "Detroit", "Oklahoma City", "Portland", "Las Vegas", "Memphis", "Louisville", "Baltimore", "Milwaukee", "Albuquerque",
        "Tucson", "Fresno", "Sacramento", "Kansas City", "Mesa", "Atlanta", "Omaha", "Colorado Springs", "Raleigh", "Miami",
        "Long Beach", "Virginia Beach", "Oakland", "Minneapolis", "Tulsa", "Arlington", "Tampa", "New Orleans", "Wichita", "Cleveland",
        "Bakersfield", "Aurora", "Anaheim", "Santa Ana", "Riverside", "Corpus Christi", "Lexington", "Stockton", "Henderson", "Saint Paul",
        "St. Louis", "Cincinnati", "Pittsburgh", "Greensboro", "Anchorage", "Plano", "Lincoln", "Orlando", "Irvine", "Newark",
        "Durham", "Chula Vista", "Toledo", "Fort Wayne", "St. Petersburg", "Laredo", "Jersey City", "Chandler", "Madison", "Lubbock",
        "Scottsdale", "Reno", "Buffalo", "Gilbert", "Glendale", "North Las Vegas", "Winston-Salem", "Chesapeake", "Norfolk", "Fremont",
        "Garland", "Irving", "Hialeah", "Richmond", "Boise", "Spokane", "Baton Rouge", "Tacoma", "San Bernardino", "Modesto",
        "Fontana", "Des Moines", "Moreno Valley", "Santa Clarita", "Fayetteville", "Birmingham", "Oxnard", "Rochester", "Port St. Lucie",
        "Grand Rapids", "Huntsville", "Salt Lake City", "Frisco", "Yonkers", "Amarillo", "Glendale", "Huntington Beach", "McKinney", "Montgomery",
        "Augusta", "Aurora", "Akron", "Little Rock", "Tempe", "Columbus", "Overland Park", "Grand Prairie", "Tallahassee", "Cape Coral"
    ]
    return random.choice(locations)

# Function to generate random vendors
def generate_vendor():
    vendors = [
        "Cisco", "Juniper", "Arista", "Palo Alto Networks", "Fortinet",
        "HP Enterprise", "Dell", "IBM", "Huawei", "Nokia",
        "Ericsson", "ZTE", "Netgear", "TP-Link", "D-Link",
        "Extreme Networks", "Ubiquiti", "Check Point", "SonicWall", "Barracuda",
        "Sophos", "F5 Networks", "Riverbed", "A10 Networks", "Citrix",
        "VMware", "Microsoft", "Google Cloud", "AWS", "Oracle",
        "Intel", "AMD", "Broadcom", "Qualcomm", "Lenovo",
        "Apple", "Samsung", "Xiaomi", "Asus", "MicroTik",
        "Brocade", "Avaya", "Alcatel-Lucent", "CommScope", "Infoblox",
        "SolarWinds", "Ruckus Wireless", "Belkin", "Blue Coat", "Ciena"
    ]
    return random.choice(vendors)

# Function to generate random admin contacts
def generate_admin_contact():
    prefixes = ["admin", "support", "user", "contact", "team"]
    domains = ["company.com", "network.com", "provider.org", "service.net", "domain.co"]
    return f"{random.choice(prefixes)}{random.randint(1, 500)}@{random.choice(domains)}"


# Function to generate random statuses
def generate_status():
    statuses = ["Active", "Inactive", "Pending", "Suspended", "Decommissioned"]
    return random.choice(statuses)

# Function to generate random priorities
def generate_priority():
    priorities = ["High", "Medium", "Low"]
    return random.choice(priorities)

# Function to generate random operational states
def generate_operational_state():
    states = ["Operational", "Maintenance", "Testing", "Outage"]
    return random.choice(states)

# Function to generate random remarks
def generate_remarks():
    remarks_list = [
        "Initial configuration", "Software upgrade", "Policy update", "Hardware installation", "Network optimization",
        "Security patching", "System maintenance", "Firmware update", "Data migration", "Performance tuning",
        "Service activation", "Configuration backup", "Access control modification", "System reboot", "Patch deployment",
        "User access setup", "Security audit", "Database restoration", "Device calibration", "Firewall rule update",
        "Backup verification", "Load balancing configuration", "Configuration tuning", "License renewal", "Network segmentation",
        "Traffic routing update", "Device registration", "Redundancy setup", "Security protocol enhancement", "Account provisioning",
        "Site inspection", "Resource allocation", "Load test completion", "Compliance check", "Bandwidth adjustment",
        "Virtual machine provisioning", "Storage expansion", "Data validation", "Connectivity testing", "User authentication setup",
        "Policy enforcement", "DNS configuration", "Application deployment", "Performance monitoring setup", "Risk assessment",
        "Server migration", "Intrusion detection configuration", "VPN setup", "Software patching", "System upgrade",
        "Scheduled maintenance", "IP address allocation", "Database indexing", "Endpoint protection update", "Disaster recovery testing",
        "Server hardening", "Traffic encryption", "Web service configuration", "Incident resolution", "Cluster expansion",
        "Environment setup", "Configuration validation", "Resource scaling"
    ]
    return random.choice(remarks_list)

# Define the number of rows
num_rows = 10000

# Generate random data
data = {
    "NE_ID": [1000 + i for i in range(1, num_rows + 1)],
    "NE_Name": [random.choice(names) for _ in range(num_rows)],
    "NE_Type": [random.choice(types) for _ in range(num_rows)],
    "NE_Version": [random.choice(versions) for _ in range(num_rows)],
    "Release_Version": [generate_random_version() for _ in range(num_rows)],
    "Board_Type": [random.choice(board_types) for _ in range(num_rows)],
    "IP_Address": [generate_ip_address(f"{base_ip}.{i // 254 + 1}") for i in range(num_rows)],
    "Subnet_Mask": ["255.255.255.0" for _ in range(num_rows)],
    "Gateway": [generate_ip_address(f"{base_ip}.{i // 254 + 1}") for i in range(num_rows)],
    "Configuration_Date": [generate_random_date(datetime(2024, 1, 1), 365) for _ in range(num_rows)],
    "Location": [generate_location() for _ in range(num_rows)],
    "Vendor": [generate_vendor() for _ in range(num_rows)],
    "Admin_Contact": [generate_admin_contact() for _ in range(num_rows)],
    "Status": [generate_status() for _ in range(num_rows)],
    "Priority": [generate_priority() for _ in range(num_rows)],
    "Operational_State": [generate_operational_state() for _ in range(num_rows)],
    "Remarks": [generate_remarks() for _ in range(num_rows)],
}

# Create a DataFrame
df = pd.DataFrame(data)

# Define the output file name
output_file = "NE_dataset.csv"

# Write to a CSV file in the desired format
with open(output_file, "w") as file:
    # Write the title row
    file.write("# Network Details,,,,,,,,,,,,,\n")
    # Write the column headers with extra commas
    file.write("NE_ID,NE_Name,NE_Type,NE_Version,Release_Version,Board_Type,IP_Address,Subnet_Mask,Gateway,Configuration_Date,Location,Vendor,Admin_Contact,Status,Priority,Operational_State,Remarks,,,,,,\n")
    # Write the data rows
    df.to_csv(file, index=False, header=False)

print(f"{num_rows} rows of data have been written to '{output_file}'")

# Download the file in Colab
from google.colab import files
files.download(output_file)
```
